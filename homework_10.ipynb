```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

plt.style.use('fivethirtyeight')
```

```python
train = pd.read_table('SouthGermanCredit.asc')
```

```python
train.head()
```

```python
array = train["laufkont laufzeit moral verw hoehe sparkont beszeit rate famges buerge wohnzeit verm alter weitkred wohn bishkred beruf pers telef gastarb kredit"]
array[:15]
```

```python
new_array = []
for i in array:
    result = list(map(int, i.split()))
    new_array.append(result)

new_array[:15]
```

```python
str_col = train.columns[0]
```

```python
train_col = str_col.split()
train_col
```

```python
data = np.array(new_array)
df = pd.DataFrame(data=data, columns=train_col)
df.head()
```

```python
from sklearn import preprocessing
```

```python
df_norm = preprocessing.normalize(df)
```

```python
df_norm.shape
```

```python
from sklearn.cluster import DBSCAN, KMeans
```

```python
kmeans = KMeans(n_clusters=2)
kmeans
```

```python
KModel = kmeans.fit(df_norm)
KModel
```

```python
KModel.labels_
```

```python
KModel.cluster_centers_
```

```python
plt.scatter(df_norm[:,0], df_norm[:,1], c=kmeans.labels_, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='*', s=200, c='red')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering Results')

plt.show()       
```

```python
dbscan = DBSCAN(eps=3, min_samples=2)
labels = dbscan.fit_predict(df_norm)
labels
```

```python
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print(f"Number of clusters found by DBSCAN: {n_clusters}")
```

```python
unique_labels = set(labels)
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        col = [0, 0, 0, 1]  
    class_member_mask = (labels == k)
    xy = df_norm[class_member_mask & core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=14,
    )

    xy = df_norm[class_member_mask & ~core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=6,
    )

plt.title(f"Estimated number of clusters: {n_clusters}")
plt.show()
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score
```

```python
kmeans_2 = KMeans(n_clusters=2, random_state=42).fit(df_norm)
kmeans_3 = KMeans(n_clusters=3, random_state=42).fit(df_norm)
kmeans_4 = KMeans(n_clusters=4, random_state=42).fit(df_norm)
kmeans_5 = KMeans(n_clusters=5, random_state=42).fit(df_norm)
```

```python
scores = {}
for name, labels in [
    ("kmeans_2", kmeans_2.labels_),
    ("kmeans_3", kmeans_3.labels_),
    ("kmeans_4", kmeans_4.labels_),
    ("kmeans_5", kmeans_5.labels_),
]:
    scores[name] = (
        silhouette_score(df_norm, labels),
        calinski_harabasz_score(df_norm, labels),
    )
```

```python
for name, (s_score, ch_score) in scores.items():
    print(f"{name}: Silhouette Score = {s_score:.2f}, Calinski-Harabasz Index = {ch_score:.2f}")
```
